```python
from utils.daod import DAOD
import cv2
from PIL import Image
import numpy as np
```

```python
arquivo = "./utils/raposa.jpg"
imagem_base = cv2.imread(arquivo)[:,:,::-1] #bgr -> rgb # OpenCV
# imagem = np.asarray(Image.open(arquivo)) PILLOW

dd = DAOD()
```

```python
# Standard image
dd.show(imagem_base)
```

```python
# Letterbox function
imagem = dd.letterbox(imagem_base, 250)
dd.show(imagem)
```

```python
# Resize function
imagem = dd.resize(imagem_base, 250)
dd.show(imagem)
```

```python
# Crop function
imagem = dd.crop(imagem_base, 100, 200, 50, 100)
dd.show(imagem)
```
